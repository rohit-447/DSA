### Q. Array is not shorted and size of array is n and after the integer ends in sequence infinity starts. ex [1,2,4,5,6,8,7, 'inf', 'inf', 'inf', 'inf']. Find the postion of first inf.

```python
#function
def linear_search(arr):
    for i in range(0, len(arr)):
        if arr[i]=='inf':
            return i



# Driver Code
arr =[1,2,3,4,5,6,7,3,4,5,63,4,5,6,'inf','inf','inf','inf','inf','inf','inf','inf','inf','inf','inf','inf']
inf_index=linear_search(arr)
print(inf_index)

#time complexity : O(n)
```

```python
#function
def binary_search(arr, x, left, right):
    if left>right:
        return -1
    mid = left + (right - left)//2
    if arr[mid]==x:
        if arr[mid-1]==x:
            return binary_search(arr, x, left, mid-1)
        else:
            return mid
            
    elif arr[mid]<x:
        return binary_search(arr, x, mid+1, right)
        
    else:
        return binary_search(arr, x, left, mid-1)
        

#driver code
arr =[1,2,3,4,5,6,7,3,4,5,63,4,5,6,float('inf'),float('inf'),float('inf'),float('inf'),float('inf'),float('inf'),float('inf'),float('inf'),float('inf'),float('inf'),]
x=float('inf')
left=0
right=len(arr)-1
inf_index=binary_search(arr, x, left, right)
print(inf_index)

#time complex : O(log(n))
```

### Two Sum

Q Find the value of the array for which the sum is 240 From the array [20,40,60,80,90,120,240].

```python
#Function
def linearsearch(arr, value):
    for i in range(0, len(arr)):
        for j in range(i+1, len(arr)):
            if arr[i]+arr[j]==value:
                return arr[i], arr[j]

#driver code
arr=[20,40,60,80,90,120,240]
value=110
result=linearsearch(arr, value)
print(result)

#time complexity : O(n^2)
```

```python
#Function
def two_sum(arr, target):
    for i in range(0, len(arr)-1):
        target_value=target-arr[i]
        idx= binary_search(arr, target_value, i+1, len(arr)-1)
        if idx!=-1:
            return arr[i], arr[idx]
            
    return None

        
def binary_search(arr, target, left, right):
    while left<=right:
        mid= left + (right-left)//2
        if arr[mid]==target:
            return mid
        elif arr[mid]<target:
            return binary_search(arr, target, mid+1, right)
        else:
            return binary_search(arr, target, left, mid-1)
            
    return -1 


#driver code
arr=[20,40,60,80,90,120,240]
target=140
result=two_sum(arr, target)
print(result)

#time complexity : O(n log(n))
```

```python
#Two pointer approch 
#function
def two_sum(arr, target, l, r):
    while l<=r:
        if arr[l]+ arr[r]==target:
            return arr[l],arr[r]
        elif arr[l]+arr[r]> target:
            r=r-1
        else:
            l=l+1


#driver code
arr=[20,40,60,80,90,120,240]
target=110
result=two_sum(arr, target, l=0, r=len(arr)-1)
print(result)

#time complexity: O(n)
```

### Q Prices of the stock are given in an arrayy. Find the max profit. 

```python
#function 
def maxprofit(arr):
    maxp=0
    for i in range (0, len(arr)):
        for j in range(i+1, len(arr)):
            if maxp<arr[j]-arr[i]:
                maxp=arr[j]-arr[i]

    return maxp
        

#time complexity O(n^2)
#space Complexity O(1)
```

```python
#function Optimised
def maxprofit(arr):
    min_price=float('inf')
    max_profit=0
    for i in range(len(arr)):
        if arr[i]<min_price:
            min_price=arr[i]
        elif arr[i]-min_price>max_profit:
            max_profit=arr[i]-min_price
    return max_profit

#time complexity: O(n)
#space complexity: O(1)
```

```python
#driver Code
arr=[4,1,5,3,6,4]
result=maxprofit(arr)
print(result)

```

### Q. 2D Array of m*n integer in each row is sorted from left to right and each row > last integer of previous row. Find integers 'x' if present return True

```python
#function
def find_array(arr, x):
    for i in range(0, len(arr)):
        for j in range(0, len(arr)+1):
            if arr[i][j]==x:
                return True
    return False

#time complexity  : O(n^2)
#space complexity : O(1)
```

```python
#function optimised
def find_array(arr, x):
    column=len(arr)
    row=len(arr[0])
    for i in range(0, column):
        if arr[i][0] <= x:
            for j in range(0, row):
                if arr[i][j]==x:
                    return True
    return False

#time complexity : O(m*n)
#space Complexity : O(1)
```

```python
#function optimised Binary search
def find_array(arr, x):
    m=len(arr)     #row
    n=len(arr[0])  #column
    low=0
    high=m*n-1
    
    while high>=low:
        mid=(high+low)//2
        row=mid//n
        column=mid%n
        if arr[row][column]==x:
            return True
        elif arr[row][column]<x:
            low=mid+1
        else:
            high = mid-1
    return False

#time complexity : O(log(m*n))
#space complexity : O(1)
```

```python
#driver code
import numpy as np
arr=[[1,3,5,7],[10,11,16,20],[23,30,34,60]]
x=1
result=find_array(arr, x)
print(result)
```

# Q. Given an array arr[], check whether it is sorted in non-decreasing order. Return true if it is sorted otherwise false.

```python
# function

#brute force
def check_sorted(arr):
    n=len(arr)
    for i in range(0, n):
        for j in range(i, n):
            if arr[i]<arr[j]:
                return False
    return True

#TC: O(n)
#SC: O(1)

#optimes
def check_sorted(arr):
    n=len(arr)
    for i in range(0, n-1):
        if arr[i]<arr[i+1]:
            return False
    return True
```

# Remove Duplicates from Sorted Array

You are given a sorted array arr[] containing positive integers. Your task is to remove all duplicate elements from this array such that each element appears only once. Return an array containing these distinct elements in the same order as they appeared.
Examples :

Input: arr[] = [2, 2, 2, 2, 2]
Output: [2]
Explanation: After removing all the duplicates only one instance of 2 will remain i.e. [2] so modified array will contains 2 at first position and you should return array containing [2] after modifying the array.
Input: arr[] = [1, 2, 4]
Output: [1, 2, 4]
Explation:  As the array does not contain any duplicates so you should return [1, 2, 4].


```python
# fun
def unique_element(arr):
    n=len(arr)
    last_idx=0
    for i in range(n-1):
        if arr[i]==arr[i+1]:
            i+=1
        elif arr[i]!=arr[i+1]:
            arr[last_idx+1], arr[i]=arr[i+1], arr[i]
            last_idx=i
    return arr

#Brute force dict
def unique_element1(arr):
    n=len(arr)
    dic={}
    for i in range(0, n):
        dic[arr[i]]=[0]

    j=0
    for key in dic:
        arr[j]=key
        j+=1
    return arr, j
```
