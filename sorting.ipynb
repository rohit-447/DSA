# Bubble Sort

```python
#Function
def bubble_sort(arr):
    n=len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j]>arr[j+1]:
                arr[j], arr[j+1]=arr[j+1], arr[j]
    return arr

#Comparsion = n^2/2
#Swaps      = n^2/2
#time Complexity = O(n^2/2 +n^2/2) = O(n^2)
#space Complexity = O(1)

#inplace sorting algo 
```

```python
#driver Code
arr=[70,20,50,30,90,5,15]
results=bubble_sort(arr)
print("Sorted array",results )
```

# Selection Sort

```python
#function
def selection_sort(arr):
    n=len(arr)
    for i in range(n):
        min_idx=i
        for j in range(i+1, n):
            if arr[j]<arr[min_idx]:
                min_idx=j
        #swap
        arr[i], arr[min_idx]=arr[min_idx], arr[i]
    return arr

#comparison: (n(n-1)/2)
#swaps : n
#time Complexity: O(n^2)
#space Complexity : O(1)
```

```python
#driver code
arr=[70,20,50,30,90,5,15]
results=selection_sort(arr)
print("Sorted array",results )
```

# Insertion  Sort


```python
#function
def insterion_sort(arr):
    n=len(arr)
    for i in range(1, len(arr)):
        j=i-1
        key=arr[i]
        while j>=0 and arr[j]>key:
            arr[j+1] =arr[j]
            j=j-1
        arr[j+1]=key
    return arr


#comparison: (n(n-1)/2)
#swaps : (n(n-1)/2)
#time Complexity: O(n^2)
#space Complexity : O(1)    
```

```python
#driver code

arr=[1,3,45,5,4,87,65,77,7,5,66]
results=insterion_sort(arr)
print("Sorted Array", results)
```
