# Bubble Sort

```python
#Function
def bubble_sort(arr):
    n=len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j]>arr[j+1]:
                arr[j], arr[j+1]=arr[j+1], arr[j]
    return arr

#Comparsion = n^2/2
#Swaps      = n^2/2
#time Complexity = O(n^2/2 +n^2/2) = O(n^2)
#space Complexity = O(1)

#inplace sorting algo 
```

```python
#driver Code
arr=[70,20,50,30,90,5,15]
results=bubble_sort(arr)
print("Sorted array",results )
```

# Selection Sort

```python
#function
def selection_sort(arr):
    n=len(arr)
    for i in range(n):
        min_idx=i
        for j in range(i+1, n):
            if arr[j]<arr[min_idx]:
                min_idx=j
        #swap
        arr[i], arr[min_idx]=arr[min_idx], arr[i]
    return arr

#comparison: (n(n-1)/2)
#swaps : n
#time Complexity: O(n^2)
#space Complexity : O(1)
```

```python
#driver code
arr=[70,20,50,30,90,5,15]
results=selection_sort(arr)
print("Sorted array",results )
```

# Insertion  Sort


```python
#function
def insterion_sort(arr):
    n=len(arr)
    for i in range(1, len(arr)):
        j=i-1
        key=arr[i]
        while j>=0 and arr[j]>key:
            arr[j+1] =arr[j]
            j=j-1
        arr[j+1]=key
    return arr


#comparison: (n(n-1)/2)
#swaps : (n(n-1)/2)
#time Complexity: O(n^2)
#space Complexity : O(1)    
```

```python
#driver code

arr=[1,3,45,5,4,87,65,77,7,5,66]
results=insterion_sort(arr)
print("Sorted Array", results)
```

# Merge Sort

```python
#merge two arrays 
#merge only works correctly if both arrays are already sorted.


def merge_array(l, r):
    result = []
    i, j = 0, 0
    n, m = len(l), len(r)

    while i < n and j < m:
        if l[i] <= r[j]:
            result.append(l[i])
            i += 1
        else:
            result.append(r[j])
            j += 1


    while i < n:
        result.append(l[i])
        i += 1

    while j < m:
        result.append(r[j])
        j += 1

    return result




def merge_sort(arr):
    if len(arr)<=1:
        return arr
    mid=len(arr)//2
    left_arr=arr[:mid]
    right_arr=arr[mid:]
    left_half=merge_sort(left_arr)
    right_half=merge_sort(right_arr)
    return merge_array(left_half, right_half)
# # Driver code
# l = [1, 3, 4, 5, 45]
# r = [5, 7, 65, 66, 77, 87]
arr=[1,3,45,5,4,87,65,77,7,5,66]
results=merge_sort(arr)
print("Sorted Array:", results)
```

# Quick sort

```python
def partitation(arr, low, high):
    pivot=arr[low]
    i, j = low, high
    while i <j:
        while arr[i]<=pivot and i<=high-1:
            i+=1
            
        while arr[j]>pivot and j >= low+1:
            j-=1
            
        if i<j:
            arr[i], arr[j]= arr[j], arr[i]
    arr[low], arr[j]= arr[j], arr[low]
    return i

def quick_sort(arr, low, high):
    if low<high:
        p_idx= partitation(arr, low, high)
        quick_sort(arr, low, p_idx-1)
        quick_sort(arr, p_idx+1, high)
        
```
