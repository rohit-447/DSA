# Bubble Sort

```python
#Function
def bubble_sort(arr):
    n=len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j]>arr[j+1]:
                arr[j], arr[j+1]=arr[j+1], arr[j]
    return arr

#Comparsion = n^2/2
#Swaps      = n^2/2
#time Complexity = O(n^2/2 +n^2/2) = O(n^2)
#space Complexity = O(1)

#inplace sorting algo 
```

```python
#driver Code
arr=[70,20,50,30,90,5,15]
results=bubble_sort(arr)
print("Sorted array",results )
```
