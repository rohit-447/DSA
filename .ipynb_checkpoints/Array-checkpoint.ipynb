# Linear Search

```python
#function
def linear_search(arr,x):
    for i in range(0, len(arr)):
        if arr[i]==x:
            return i
    return -1



#driver code
arr=[10,20,30,40,50,60,70,80]
x=600
result=linear_search(arr, x)
print(result)

#time Com : O(1)
```

# Binary Search

```python
#function
def binary_search(arr, x , left, right):
    while left<=right:
        mid=left + (right-left) //2

        #mid search
        if arr[mid]==x:
            return mid

        #right search
        elif arr[mid]<x:
            return binary_search(arr, x, mid+1, right)

        #left search
        else:
        #elif arr[mid]>x:
            return binary_search(arr, x, left, mid-1)

    return -1


#driver code
arr=[10,20,30,40,50,60,70,80]
arr.sort()
x=60
left=0
right=len(arr)-1
result = binary_search(arr, x, left, right)
print(result)

#time complex: O(log(n))
#space complex: O(1)
```
