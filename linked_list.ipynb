### At beginning

```python

#Class node
class Node():
    def __init__(self, data):
        self.data=data
        self.next=None



#class Linked List        
class Linkedlist():
    def __init__(self, data):
        self.head=None


    #intert the node at beginning
    def insertAtbeginning(self, new_data):
        new_node=Node(new_data)
        new_node.next=self.head
        self.head=new_node

        
    #print the linked List
    def printLL(self):
        temp=self.head
        while temp:
            print(temp.data + " ", end=' ')
            temp=temp.next
```

```python
ll=Linkedlist()
ll.insertAtbeginning(10)
ll.insertAtbeginning(20)
ll.insertAtbeginning(30)
ll.printLL()
```
