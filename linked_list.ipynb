### At beginning

```python

#Class node
class Node():
    def __init__(self, data):
        self.data=data
        self.next=None



#class Linked List        
class Linkedlist():
    def __init__(self, data):
        self.head=None


    #intert the node at beginning
    def insertAtbeginning(self, new_data):
        new_node=Node(new_data)
        new_node.next=self.head
        self.head=new_node

        
    #print the linked List
    def printLL(self):
        temp=self.head
        while temp:
            print(temp.data + " ", end=' ')
            temp=temp.next
```

```python
ll=Linkedlist()
ll.insertAtbeginning(10)
ll.insertAtbeginning(20)
ll.insertAtbeginning(30)
ll.printLL()
```

### At End

```python
class Node():
    def __init__(self, data):
        self.data=data
        self.next=None

class linkedlist():
    def __init__(self):
        self.head=None

    def insert_at_end(self, new_data):
        new_node=Node(new_data)
        if self.head is None:
            self.head=new_node
            return
            
        temp=self.head
        while temp.next:
            temp=temp.next

        temp.next=new_node


    def print_ll(self):
        temp=self.head
        while temp:
            print(temp.data, end=' ')
            temp=temp.next
```

```python
ll_at_end=linkedlist()
ll_at_end.insert_at_end(1)
ll_at_end.insert_at_end(2)
ll_at_end.insert_at_end(3)
ll_at_end.insert_at_end(4)
ll_at_end.print_ll()
```

### At after node

```python
class Node():
    def __init__(self, data):
        self.data=data
        self.next=None


class Linked_list():
    def __init__(self):
        self.head=None

    def insertAtbeginning(self, new_data):
        new_node=Node(new_data)
        new_node.next=self.head
        self.head=new_node

    def insertion_after_Node(self, prev_node, new_data):
        new_node=Node(new_data)

        if prev_node is None:
            return -1

        new_node=Node(new_data)
        new_node.next=prev_node.next
        prev_node.next=new_node
        
         
    #print the linked List
    def printLL(self):
        temp=self.head
        while temp:
            print(temp.data, end=' ')
            temp=temp.next
```

```python
ll=Linked_list()
ll.insertAtbeginning(10)
ll.insertAtbeginning(20)
ll.insertAtbeginning(30)
ll.printLL()
print("\nAfter Insertion")
ll.insertion_after_Node(ll.head.next,25)
ll.printLL()
```
