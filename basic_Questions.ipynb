```python
#sum of digits
digit=1234
sum_d=0
while digit!=0:
    digit1=digit%10
    digit=digit//10
    sum_d=sum_d+digit1
print(sum_d)
```

```python
# Remove spaces from a string
i_string='p y t h o n'
i_string=i_string.replace(' ', '')
print(i_string)
```

```python
# Replace all vowels in a string with *
i_string='python'
for char in i_string:
    if char =='a' or char =='e' or char =='i' or char =='o' or char =='u':
        i_string=i_string.replace(char,'*')
print(i_string)
```

```python
# Count occurrences of a character in a string
i_string='Hello'
occ_word='l'
count=0
for i in i_string:
    if i==occ_word:
       count=count+1
print(count)
```

```python
# int pallendrom or not
a=22
rev=0

copy_d=a
while a!=0:
    digit=a%10
    rev=rev*10+digit
    a=a//10

print(rev)
if copy_d==rev:
    print(True)
else:
    print(False)
```
