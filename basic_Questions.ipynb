```python
# Find no. of digits
import math

#way 1:
def no_digit_1(digit):
    count=0
    if digit==0:
        return 1
    while digit>0:
        digit=digit//10
        count+=1
    return count


#way 2
def no_digit_2(digit):
    count=0
    if digit==0:
        return 1
    count=int(math.log10(abs(digit))+1)
    return count


n=11
out1=no_digit_1(n)
out2=no_digit_2(n)
print(out1, out2)
```

```python
#Armstong no.
def armstrong(digit):
    digit_c=digit
    pan=0
    while digit_c>0:
        reminder=digit_c%10
        pan=reminder**3+pan
        digit_c=digit_c//10

    if pan==digit:
        return True
    else:
        return False

result=armstrong(153)
print(result)
        
```

```python
# Prime no
def find_prime(digit):
    if digit <=1:
        return False
        
    for i in range(2, int((digit)**0.5)+1):
        if digit%i==0:
            return False 
       
    return True

digit=2
result=find_prime(digit)
print(result)
```

```python
# prime no. upto N
def prime_n(n):
    c = 0
    for num in range(2, n + 1):
        is_prime = True
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print(num, end=' ')
            c += 1
    return c


n = 100
result = prime_n(100)
print("Total primes:", result)
```

```python
#prime no. in a range
def primeRange(M,N):
    for i in range(M, N+1):
        is_prime=True
        for num in range(2, int(i**0.5)+1):
            if i%num==0:
                is_prime=False
                break
        if is_prime:
                if i==1:
                    continue
                print(i, end=' ')
                
             
primeRange(1,10)
```
